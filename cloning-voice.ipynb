```python
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F
import IPython
```

```python
from tortoise.api import TextToSpeech
from tortoise import utils
```

```python
tts = TextToSpeech()
```

```python
from huggingface_hub import snapshot_download
local_dir = "/wav_examples"
snapshot_download(repo_id="styalai/wav-examples", local_dir=local_dir, repo_type="dataset")
```

```python
from os import listdir
data = "/kaggle/input/wav-examples"
clips_paths = listdir(data)
reference_clips = [utils.audio.load_audio(data+"/"+p, 22050) for p in clips_paths]
```

```python
# ultra_fast, fast
audio = tts.tts_with_preset("For this training, I’ll train two models with the same base, the same time", voice_samples=reference_clips, preset='high_quality')
```

```python
audio.shape
```

```python
torchaudio.save(f'generated-high_quality.wav', torch.squeeze(audio, 0).cpu(), 24000)
```
